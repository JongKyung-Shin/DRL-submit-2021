```python
import numpy as np
import json
import pandas as pd
import os
import sys
__file__ = "state_generator.py"
dirname = os.path.dirname(os.path.abspath(os.path.dirname(__file__)))
sys.path.append(dirname)

with open(dirname + "/Data/json_file/purchase_202001_ver4.json") as json_file:
    stream_data = json.load(json_file)

key_list = list(stream_data.keys())
```

```python
def drop_duplicate(stream_data):
    for key in stream_data.keys():
        new_list = []
        good_list = stream_data[key]["purchase_list"]
        for good in good_list:
            if good not in new_list:
                new_list.append(good)
        stream_data[key]["purchase_list"] = new_list
    return stream_data
```

```python
drop_data= drop_duplicate(stream_data)
```

```python
key_3_list = []
key_5_list = []
key_10_list = []
key_20_list = []
for key in key_list:
    goods_list = drop_data[key]["purchase_list"]
    if len(goods_list) == 3:
        key_3_list.append(key)
    elif len(goods_list) == 5:
        key_5_list.append(key)
    elif len(goods_list) == 10:
        key_10_list.append(key)
    elif len(goods_list) == 20:
        key_20_list.append(key)
    else:
        pass
```

```python
key_df = pd.DataFrame(columns= ["key"])
key_df["key"] = key_3_list
key_df.to_csv(dirname + "/Data/key_3_list.csv", index = False)
```

```python
# 전체 상품 리스트 추출
def goods_get(key_list, stream_data):
    total_goods_list = []
    for key in key_list:
        good_list = stream_data[key]["purchase_list"]
        for good in good_list:
            total_goods_list.append(good["상품코드"])
    #total_goods_list = list(set(total_goods_list))
    return total_goods_list
```

```python
good_3_list = goods_get(key_3_list, stream_data)
good_5_list = goods_get(key_5_list, stream_data)
good_10_list = goods_get(key_10_list, stream_data)
```

```python
# 상품별 빈도 수 확인
def check_frequency(good_list):
    freq_dic = {}
    for good in good_list:
        if good not in freq_dic.keys():
            freq_dic[good] = 1
        else:
            freq_dic[good] += 1
    
    items = {k: v for k, v in sorted(freq_dic.items(), key=lambda item: item[1], reverse = True)}
   
    return items
```

```python
freq_3_dic = check_frequency(good_3_list)
freq_5_dic = check_frequency(good_5_list)
freq_10_dic = check_frequency(good_10_list)
```

```python
#서로 다른 sequence 수 확인
def unique_sequence_count(stream_data, num_key):
    sequence_dic = {}
    for key in stream_data.keys():
        name_list = []
        good_list = stream_data[key]["purchase_list"]
        if len(good_list) == num_key:
            for good in good_list:
                name_list.append(good["상품명"])
            if str(name_list) not in sequence_dic.keys():
                sequence_dic[str(name_list)] = 1
            else:
                sequence_dic[str(name_list)] += 1
    return sequence_dic
```
